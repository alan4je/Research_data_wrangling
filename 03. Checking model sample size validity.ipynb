```python
segments = []
for i in np.unique(np.array(df.UID)):
    sdf = df[df.UID == i]
    years = np.array(sdf.sYear)
    rehab = np.array(sdf.loc[:, 'Major Rehab Year'])[0]
    if rehab in years:
        if (rehab -1) in years:
            if (rehab+1) in years:
                if (rehab+2) in years:
                    segments.append(i)
```

```python
segmentsCPM = []
for i in np.unique(np.array(df.UID)):
    sdf = df[df.UID == i]
    years = np.array(sdf.sYear)
    rehab = np.array(sdf.loc[:, 'Major Rehab Year'])[0]
    if rehab in years:
            if (rehab+1) in years:
                if (rehab+2) in years:
                    if (rehab+3) in years:
                        if (rehab+4) in years:
                            segmentsCPM.append(i)
```
