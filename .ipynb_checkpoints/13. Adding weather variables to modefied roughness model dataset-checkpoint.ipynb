```python
import pandas as pd
import numpy as np
```

```python
ppt = pd.read_csv('../FinalDatasets/ppt.csv')
maxt = pd.read_csv('../FinalDatasets/MaxT.csv')
mint = pd.read_csv('../FinalDatasets/MinT.csv')
esals =pd.read_csv('../FinalDatasets/ESALs.csv')
rdat = pd.read_csv('../modelDatasets/rmdata1988_1999.csv')
```

```python
print len(np.unique(np.array(esals.ID)))*33
```

```python
selected_ids = np.unique(np.array(ppt.ID))
esals = esals[esals.ID.isin(selected_ids)]
```

```python
esals = esals.drop_duplicates(subset = ['ID'])
ppt  = ppt.drop_duplicates(subset = ['ID'])
mint = mint.drop_duplicates(subset = ['ID'])
maxt = maxt.drop_duplicates(subset = ['ID'])
```

```python
print len(esals)
print len(ppt)
print len(mint)
print len(maxt)
print len(rdat)
```

```python
esals.index = esals.ID
ppt.index = ppt.ID
mint.index = mint.ID
maxt.index = maxt.ID
```

```python
def addVariable(row, df):
    index = int(row['UID'])
    year  = str(row['sYear'])
    return df.loc[index, year]
def addPPT(row):
    index = int(row['UID'])
    year  = str(row['sYear'])
    return ppt.loc[index, year]
def addMinT(row):
    index = int(row['UID'])
    year  = str(row['sYear'])
    return mint.loc[index, year]
def addMaxT(row):
    index = int(row['UID'])
    year  = str(row['sYear'])
    return maxt.loc[index, year]
```

```python
rdat.loc[:, 'esals'] = rdat.apply(addVariable, axis = 1, args = (esals,))
rdat.loc[:, 'ppt'] = rdat.apply(addPPT, axis = 1)
rdat.loc[:, 'mint'] = rdat.apply(addMinT, axis = 1)
rdat.loc[:, 'maxt'] = rdat.apply(addMaxT, axis = 1)
```
